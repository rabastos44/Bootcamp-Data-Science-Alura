# Desafio: adicionar uma coluna mostrando a região de cada estado

```
estados = usaveis.index

regiao = {'1':'Norte',
          '2':'Nordeste',
          '3':'Sudeste',
          '4':'Sul',
          '5':'Centro-Oeste'}

col_regiao = []

for a in estados:
  col_regiao.append(regiao[a[0]])

usaveis.insert(0,'Região',col_regiao)
```

```
usaveis.head()
```

```
usaveis = usaveis.drop("Região", axis=1)
```

# Aula 3.1

```
usaveis["Total"] = usaveis.sum(axis=1) # [0,1,2] ["SP"]
usaveis.head()
```

```
ordenado_por_total = usaveis.sort_values("Total", ascending=False)
ordenado_por_total = ordenado_por_total.drop("Total", axis=1)
ordenado_por_total.head()
```

# Desafio: deixar o gráfico mais bonito. A mensagem mais clara.

```
ax = ordenado_por_total.head(5).T.plot(figsize=(12,6))
ax.legend(bbox_to_anchor=(1,.5),ncol=1,loc='upper left')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos de internação do Sus - Estados que mais Gastão")
plt.set_cmap(cm.jet)
plt.show()  
```

#Desafio: adicione seu estado aos 5 estados existentes e plote
Adicionando Pernambuco

```
Top5Estados=ordenado_por_total.index[:5].tolist()
Top5Estados.append('26 Pernambuco')
```

```
ax = ordenado_por_total.T[Top5Estados].plot(figsize=(12,6))
ax.legend(bbox_to_anchor=(1,.5),ncol=1,loc='upper left')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}/"))
plt.title("Gastos de internação do Sus - Estados que mais Gastão")
plt.set_cmap(cm.jet)
plt.show()  
```
